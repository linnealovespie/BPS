```python
summary_stats_header = [
    'CompanyName',
    'NumBuildings',
    'TotalEmissions(metricTons)',
    'AverageEmissionsPerBuilding',
    'TotalElectricityUse',
    'AverageElectricityUse',
    'TotalGasUse',
    'AverageGasUse',
    'AverageGHGEmissionsIntensity'
]
```

```python
def get_summary_stats(building_df, name):
    return [
        name,
        len(building_df),
        building_df['TotalGHGEmissions__metric_tons_'].sum(),
        building_df['TotalGHGEmissions__metric_tons_'].mean(),
        building_df['Electricity_kBtu_'].sum(),
        building_df['Electricity_kBtu_'].mean(),
        building_df['NaturalGas_kBtu_'].sum(),
        building_df['NaturalGas_kBtu_'].mean(),
        building_df['GHGEmissionsIntensity'].mean()
    ]
```

```python
company_buildings = [[google_buildings, 'Google'], [tableau_buildings, 'Tableau'], [facebook_buildings, 'Facebook'], [amazon_buildings, 'Amazon']]
summary_stats_matrix = [get_summary_stats(company[0], company[1]) for company in company_buildings]
```

```python
tech_cos_comparison = pd.DataFrame.from_records(summary_stats_matrix, columns = summary_stats_header)
```
