```python
# These numbers are deliberately conservative
# ['storeOSEid', 'segments'], floor area calculated as 1/segments

groceries_and_proportions = [
    [50096, 37, 'Grocery - Amazon Go'],
    [50476, 17, 'Grocery - Amazon Go'],
    [640, 63, 'Grocery - Amazon Fresh'],
    [516, 1, 'Grocery - Amazon Fresh'],
    [50460, 13, 'Grocery - Whole Foods'],
    [700, 1, 'Grocery - Whole Foods'],
    [132, 2,'Grocery - Whole Foods'],
    [50259, 7, 'Grocery - Whole Foods']
]
```

```python
groceries = []

grocery_headers = [
    'OSEBuildingID', 
    'TaxParcelIdentificationNumber',
    'Address',
    'PropertyGFATotal',
    'TotalGHGEmissions__metric_tons_',
    'PortionOfTotalBuilding',
    'StoreGFATotal',
#     'StoreTotalGHGEmissions__metric_tons',
    'SiteEnergyUse_kBtu_',
#     'StoreTotalSiteEnergyUse_kBtu',
    'NaturalGas_kBtu_',
    'Electricity_kBtu_',
    'SteamUse_kBtu_'
]
```

```python
groceries = []
for store in groceries_and_proportions:
    data = all_buildings_dataframe[all_buildings_dataframe['OSEBuildingID'] == store[0]].iloc[0]
    groceries.append([
        data['OSEBuildingID'], # 'OSEBuildingID', 
        data['TaxParcelIdentificationNumber'], #'TaxParcelIdentificationNumber'
        data['Address'], # 'Address'
        data['PropertyGFATotal'], #'Address'
        data['TotalGHGEmissions__metric_tons_'], # 'TotalGHGEmissions__metric_tons_'
        store[1], # 'PortionOfTotalBuilding'
        data['PropertyGFATotal'] / float(store[1]),
#         data['TotalGHGEmissions__metric_tons_'] / float(store[1]),
        data['SiteEnergyUse_kBtu_'],
#         data['SiteEnergyUse_kBtu_'] / float(store[1]),
        data['NaturalGas_kBtu_'],
        data['Electricity_kBtu_'],
        data['SteamUse_kBtu_']
    ])
```

```python
grocery_df = pd.DataFrame(data=groceries, columns=grocery_headers)
```

```python
grocery_df['Owner'] = 'AMAZON'
amazon_grocery_df = grocery_df #because grocery_gf is a terrible name
grocery_df.head()
```

```python
# conservative estimate of GHGE in metric tons
# based on EPA: https://www.epa.gov/sites/default/files/documents/gc_averagestoreprofile_final_june_2011_revised_1.pdf

def estimate_ghge_for_grocery(sq_ft):
    avg_co2_fac = 1383 
    avg_ref_fac = 1556
    return ((sq_ft / 46000.0) * avg_co2_fac) + ((sq_ft / 46000) * avg_ref_fac)
```

```python
# grocery_df['StoreNaturalGas_kBtu'] = grocery_df['NaturalGas_kBtu_'] / grocery_df['PortionOfTotalBuilding']
# grocery_df['StoreElectricity_kBtu'] = grocery_df['Electricity_kBtu_'] / grocery_df['PortionOfTotalBuilding']
# grocery_df['StoreSteamUse_kBtu'] = grocery_df['SteamUse_kBtu_'] / grocery_df['PortionOfTotalBuilding']
```

```python
grocery_df['EstStoreGHGE_metric_tons'] = estimate_ghge_for_grocery(grocery_df['StoreGFATotal'])
```

##### Whoa, are the other supermarkets under-reporting??

```python
# back of the envelope for Amazon
# our numbers for amazon were calculated with electric and r-404a emissions: https://www.epa.gov/sites/default/files/documents/gc_averagestoreprofile_final_june_2011_revised_1.pdf
# maybe the other grocers didn't include that?

avg_supermarket = 46000
metric_tons_co2_for_avg_supermarket = 1383
num_avg_amazon_markets = 473258 / avg_supermarket
amazon_markets_ghge = num_avg_amazon_markets * metric_tons_co2_for_avg_supermarket
print('Estimated electric-only GHGE impact of all Amazon supermarkets')
print(amazon_markets_ghge)
```

Amazon has nearly five times the GHGE for its grocery stores compared to other grocery stores (counting electricity only)

Let's look at QFC, the largest grocery by total area:

```python
num_avg_qfc_markets = 653745 / avg_supermarket
qfc_markets_ghge = num_avg_qfc_markets * metric_tons_co2_for_avg_supermarket
print('Estimated electric-only GHGE impact of all QFC supermarkets')
print(qfc_markets_ghge)
```

So QFC would be expected to produce at least 20k metric tons of GHGE from electricity alone. And yet they produce less than 2k, allegedly!

```python
all_non_amazon_grocery[all_non_amazon_grocery['Owner'] == 'QFC']['TotalGHGEmissions__metric_tons_'].sum()
```
