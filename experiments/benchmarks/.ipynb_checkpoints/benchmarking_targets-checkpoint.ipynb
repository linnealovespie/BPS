Let's try decreasing the acceptable levels. 

Assumptions:
- if a decrease percent would be < 0, we will list it as 0 (no change)
- if a decrease percent would be >= 1, we will list it as 1
- if we decrease the allowed GHGE intensity, we calculate the new allowed level as (1 - percent_decrease) * current_level

Calculations:
- new GHGE intensity for each benchmark year
- amount of carbon produces in first year of each benchmark period

```python
emissions_benchmarks_with_current_ghge.head()
```

```python
# we need the total sq ft/category so we can calculate total emissions

current_emissions_temp_df = emissions_by_data_type[['LegislationPropertyType', 'totalSqFt']]

emissions_benchmarks_with_current_ghge = pd.merge(emissions_benchmarks_with_current_ghge,
        current_emissions_temp_df,
        on='LegislationPropertyType',
        how='left'
        )
```

```python
# we'll drop the first two categories, since they don't map to our building data

emissions_benchmarks_with_current_ghge = emissions_benchmarks_with_current_ghge.drop([0, 1])
emissions_benchmarks_with_current_ghge.reset_index(drop=True, inplace=True)
```

```python
emissions_benchmarks_with_current_ghge.columns
```

```python
def reduction_factor(percent_emissions_decrease):
    return (100 - percent_emissions_decrease) / 100

def get_emissions_estimates_df(percent_emissions_decrease):
    red_fac = reduction_factor(percent_emissions_decrease)
    
    decrease_df = pd.DataFrame({
        'Building Type': emissions_benchmarks_with_current_ghge['LegislationPropertyType'],
        'Additional Emissions Intensity Decrease Percent': percent_emissions_decrease,
        'Total Sq Footage': emissions_benchmarks_with_current_ghge['totalSqFt'],
        '2020 Average GHGE Intensity kg^2/sq ft': emissions_benchmarks_with_current_ghge['2020AvgGHGEIntensity'],
        '2024-2027 Emissions (kg^2)': emissions_benchmarks_with_current_ghge['2020AvgGHGEIntensity'] * emissions_benchmarks_with_current_ghge['totalSqFt'] * 3,
        '2027-2030 GHGE Intensity (kg^2/sq ft)': emissions_benchmarks_with_current_ghge['2027-2030'] * red_fac,
        '2027-2030 GHGE Emissions (kg^2)': emissions_benchmarks_with_current_ghge['2027-2030'] * red_fac * emissions_benchmarks_with_current_ghge['totalSqFt'] * 4,
        '2031-2035 GHGE Intensity (kg^2/sq ft)': emissions_benchmarks_with_current_ghge['2031-2035'] * red_fac,
        '2031-2035 GHGE Emissions (kg^2)': emissions_benchmarks_with_current_ghge['2031-2035'] * red_fac * emissions_benchmarks_with_current_ghge['totalSqFt'] * 5,
        '2036-2040 GHGE Intensity (kg^2/sq ft)': emissions_benchmarks_with_current_ghge['2036-2040'] * red_fac,
        '2036-2040 Emissions (kg^2)': emissions_benchmarks_with_current_ghge['2036-2040'] * red_fac * emissions_benchmarks_with_current_ghge['totalSqFt'] * 5,
        '2041-2045 GHGE Intensity (kg^2/sq ft)': emissions_benchmarks_with_current_ghge['2041-2045'] * red_fac,
        '2041-2045 Emissions (kg^2)': emissions_benchmarks_with_current_ghge['2041-2045'] * red_fac * emissions_benchmarks_with_current_ghge['totalSqFt'] * 5,
        '2046- GHGE Intensity (kg^2/sq ft)': emissions_benchmarks_with_current_ghge['2046-'] * red_fac,
        '2046- Emissions (kg^2)': emissions_benchmarks_with_current_ghge['2046-'] * red_fac * emissions_benchmarks_with_current_ghge['totalSqFt'] * 1 # these will all be 0 because GHGE intensity = 0 for 2046
    })
    
    all_types_row = {
        'Building Type': 'All Buildings',
        'Additional Emissions Intensity Decrease Percent': percent_emissions_decrease,
        'Total Sq Footage': emissions_benchmarks_with_current_ghge['totalSqFt'].sum(),
        '2020 Average GHGE Intensity kg^2/sq ft': emissions_benchmarks_with_current_ghge['2020AvgGHGEIntensity'].mean(),
        '2024-2027 Emissions (kg^2)': emissions_benchmarks_with_current_ghge['2020AvgGHGEIntensity'].mean() * emissions_benchmarks_with_current_ghge['totalSqFt'].sum() * 3,
        '2027-2030 GHGE Intensity (kg^2/sq ft)': emissions_benchmarks_with_current_ghge['2027-2030'].mean() * red_fac,
        '2027-2030 GHGE Emissions (kg^2)': emissions_benchmarks_with_current_ghge['2027-2030'].mean() * red_fac * emissions_benchmarks_with_current_ghge['totalSqFt'].sum() * 4,
        '2031-2035 GHGE Intensity (kg^2/sq ft)': emissions_benchmarks_with_current_ghge['2031-2035'].mean() * red_fac,
        '2031-2035 GHGE Emissions (kg^2)': emissions_benchmarks_with_current_ghge['2031-2035'].mean() * red_fac * emissions_benchmarks_with_current_ghge['totalSqFt'].sum() * 5,
        '2036-2040 GHGE Intensity (kg^2/sq ft)': emissions_benchmarks_with_current_ghge['2036-2040'].mean() * red_fac,
        '2036-2040 Emissions (kg^2)': emissions_benchmarks_with_current_ghge['2036-2040'].mean() * red_fac * emissions_benchmarks_with_current_ghge['totalSqFt'].sum() * 5,
        '2041-2045 GHGE Intensity (kg^2/sq ft)': emissions_benchmarks_with_current_ghge['2041-2045'].mean() * red_fac,
        '2041-2045 Emissions (kg^2)': emissions_benchmarks_with_current_ghge['2041-2045'].mean() * red_fac * emissions_benchmarks_with_current_ghge['totalSqFt'].sum() * 5,
        '2046- GHGE Intensity (kg^2/sq ft)': emissions_benchmarks_with_current_ghge['2046-'].mean() * red_fac,
        '2046- Emissions (kg^2)': emissions_benchmarks_with_current_ghge['2046-'].mean() * red_fac * emissions_benchmarks_with_current_ghge['totalSqFt'].sum() * 1 # these will all be 0 because GHGE intensity = 0 for 2046
    }
    
    return decrease_df.append(all_types_row, ignore_index=True)
```

```python
# test method 

get_emissions_estimates_df(5)
```

```python
# run get_emissions_estimates_df for all values of decrease -- 0-.95 inclusive

all_decrease_projections = pd.DataFrame()

for i in range(0, 101, 5):
    print(i)
    all_decrease_projections = all_decrease_projections.append(get_emissions_estimates_df(i), ignore_index=True)
```
